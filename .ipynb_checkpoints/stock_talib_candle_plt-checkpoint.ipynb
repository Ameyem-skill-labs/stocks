```python
# %reset
from nsefolder2pickle import NseFolder2Pickle; 
import numpy as np
import pandas as pd

from nse_pickle_fns import *
import os

# import pandas_datareader as datareader
%matplotlib inline
import matplotlib.pyplot as plt
import datetime
#from matplotlib.finance import candlestick_ohlc
from mpl_finance import candlestick_ohlc
import matplotlib.dates as mdates

# from talib import abstract
from talib.abstract import *
symbols_w=load_pkl(pickle_path+'symbols_to_work.pkl')
```

```python
def getstock_dict(symbol,datalimit):
    stock_array=np.array(load_pkl(pickle_path+symbol+'.pkl'))
    stock_array=stock_array[-datalimit:,:]
    return input_arrays = {
        'date': mdates.date2num(stock_array[:,7]).astype(float),
        'open': stock_array[:,0].astype(float),
        'high': stock_array[:,1].astype(float),
        'low': stock_array[:,2].astype(float),
        'close': stock_array[:,3].astype(float),
        'volume': stock_array[:,4].astype(float),
        'ntrades': stock_array[:,8].astype(float)
    }
```

```python
from matplotlib.gridspec import GridSpec
def stock_plt(symbol,datalimit):
    cols=[7,0,1,2,3]

    stock_array=np.array(load_pkl(pickle_path+symbol+'.pkl'))
    stock_array=stock_array[-datalimit:,:]
    clean_stock=stock_array[:,cols]
    dates=clean_stock[:,0].copy()
    clean_stock[:,0]=mdates.date2num(clean_stock[:,0])
    input_arrays = {
        'date': stock_array[:,7],
        'open': stock_array[:,0].astype(float),
        'high': stock_array[:,1].astype(float),
        'low': stock_array[:,2].astype(float),
        'close': stock_array[:,3].astype(float),
        'volume': stock_array[:,4].astype(float),
        'ntrades': stock_array[:,8].astype(float)
    }

    # sma = abstract.SMA
    # sma = abstract.Function('sma')
    output = SMA(input_arrays, timeperiod=10) # calculate on close prices by default
    # output = SMA(input_arrays, timeperiod=25, price='open') # calculate on opens
    upper, middle, lower = BBANDS(input_arrays, 20, 2, 2)
    slowk, slowd = STOCH(input_arrays, 5, 3, 0, 3, 0) # uses high, low, close by default
    slowk, slowd = STOCH(input_arrays, 5, 3, 0, 3, 0, prices=['high', 'low', 'open'])


    f1, ax = plt.subplots(2, sharex=True,figsize = (20,10))
    gs = GridSpec(3, 3)
    
    ax[0] = plt.subplot(gs[0:2, :])
    # identical to ax1 = plt.subplot(gs.new_subplotspec((0,0), colspan=3))
    ax[1] = plt.subplot(gs[-1,:])
    # plot the candlesticks
    candlestick_ohlc(ax[0], clean_stock, width=1, colorup='green', colordown='red')
    ax[0].xaxis.set_major_formatter(mdates.DateFormatter('%d-%b-%y'))

    # plot the moving average lines
    ax[0].plot(dates, output, color = 'blue', label = 'ma25')
    ax[0].plot(dates, upper, color = 'red', label = 'upper')
    ax[0].plot(dates, middle, color = 'green', label = 'middle')
    ax[0].plot(dates, lower, color = 'red', label = 'lower')

    ax[1].plot(dates, slowk, color = 'purple', label = 'slowk')
    ax[1].plot(dates, slowd, color = 'green', label = 'slowd')

    # other parameters
    ax[0].grid(True)
    ax[0].legend()
#     plt.title('Stock')
    ax[0].set_title(r'Technical analysis of '+symbol)

    plt.show()
```

```python

datalimit=500
stk_id=10
columns =['Open','High','Low','Close','Last','Volume','Value','Date','NTrades']
symbol=symbols_w[stk_id]

stock_plt(symbol,datalimit)
```
